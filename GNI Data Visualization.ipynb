```python
import matplotlib.pyplot as plt
```

```python
data.groupby('State/UnionTerritory')['Deaths'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color=['r','b']);
plt.title('Top States with Reported Cases of Deaths')
plt.ylabel('Number of Deaths')
plt.xlabel('State Territory')
plt.show();
```

```python
# Histogram of confirmed cases
plt.figure(figsize=(8, 5))
plt.hist(data['Confirmed'], bins=30, color='blue', edgecolor='black')
plt.title('Distribution of Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.show()
```

# Seaborn

```python
import seaborn as sns

# Heatmap for correlation
plt.figure(figsize=(10, 6))
sns.heatmap(data[['Confirmed',  'Cured', 'Deaths']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Boxplot for confirmed cases by state
plt.figure(figsize=(15, 6))
sns.boxplot(x='State/UnionTerritory', y='Confirmed', data=data)
plt.xticks(rotation=90)
plt.title('Confirmed Cases by State/Union Territory')
plt.show()
```

```python

# Pairplot for numerical variables
sns.pairplot(data[['Confirmed', 'Cured', 'Deaths']])
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.scatter(data.satisfaction_level, data.average_montly_hours, color='m')

plt.xlabel('Satisfaction level')
plt.ylabel('average montly hours')

plt.title('Satisfaction level vs average montly hour');
```

```python
plt.scatter(data.satisfaction_level, data.left);
```

```python
sal=data.salary.value_counts(normalize=True)*100
sal
```

```python
salary_counts = data['salary'].value_counts()
explode=[0.01,0.01,0.0]
plt.pie(salary_counts, labels=salary_counts.index, autopct='%1.1f%%', explode=explode,startangle=90)
plt.title('Salary Distribution')
plt.show()
```
